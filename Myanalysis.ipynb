### Import Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
covidvaccine = pd.read_csv("COVIDVaccineCoverage.csv")

covidvaccine.head()
```

### Data Wrangling
#### Renaming Variables


```python
covidvaccine.rename({'Black':'Vaccinated Black','South Asian':'Vaccinated South Asian','White':'Vaccinated White'}, axis=1, inplace=True)
covidvaccine
```

### Filtering Data
#### Show only colums I need

```python
Vaccinated_by_Race = covidvaccine[["Vaccinated Black","Vaccinated South Asian", "Vaccinated White","Age Range","Covid Vacc Date"]]
Vaccinated_by_Race.head()
```

* This information shows that Vaccinated White persons are consistantly the highest vaccinated people

 #### Data Aggragation

```python
Vaccinated_by_Race.groupby('Age Range').sum()
```

* This information shows that Black persons in the age range of 50-54, 55-59,& 80+ are least vaccinated

```python
Vaccinated_by_Race.groupby('Covid Vacc Date').sum()
```

* September dates can possibly be scheduled appointments

```python
Vaccinated_by_Race.corr()
Vaccinated_by_Race
```

*These colums show that Black person are the least vaccinated in age range of 50-54 & 80+

#### Data Visualization

```python
Vaccinated_by_Race.groupby(['Age Range'])['Vaccinated Black','Vaccinated South Asian','Vaccinated White'].count().plot(kind='bar')
```
